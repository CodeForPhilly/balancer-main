```python
import asyncio
from gitingest import ingest_async
import nest_asyncio

# Apply nest_asyncio to allow nested event loops
nest_asyncio.apply()

async def main():
    # Ingest the data
    summary, tree, content = await ingest_async(r"..\llm-backend")

    with open("summary.txt", "w", encoding="utf-8") as summary_file:
        summary_file.write(str(summary))
    with open("tree.txt", "w", encoding="utf-8") as tree_file:
        tree_file.write(str(tree))
    with open("content.txt", "w", encoding="utf-8") as content_file:
        content_file.write(str(content))

# Run the async main function in the running event loop
asyncio.run(main())
```
